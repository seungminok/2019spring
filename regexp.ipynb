>variable이 list이구나 를 알아야한다. 하나 하나는 str이지만, 전체는 아님. 

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
```

str의 처음과 끝은 명시가 되어있다. 

0-9 + 플러스는 앞에 있는게 통째로 반복된다. 숫자가 뭐가 와도 된다. 하나 이상의 숫자가 나오는 것을 의미. 

그냥 점을 쓰면 캐릭터 하나지만, 그냥 소수점포인트를 나타내고 싶으면, backslash로 표현한다. 

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
```

알파벳은 하나 이상 나오고, 실제 달러가 나오는 걸 원해서 /$ 를 쓴다.

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
print(wordlist[:10])
```

str의 처음이 숫자로 시작되고, {} 몇번 반복한다는 뜻

0-9까지의 숫자가 4번 반복한다 -> 네자리수를 찾아라

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
```

숫자가 + 하나 이상 나오고, dash 그 자체에다가 영어가 세자리 **또는** 다섯자리가 반복되어라. 

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
```

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

|(bar)는 or의 의미. ed로 끝나거나, ing로 끝나는 것을 찾아줌. 

```python
wordlist = [w for w in wsj if re.search('(ed|ing)$', w)]
print(wordlist[:10])
```

```python
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}',word))
fdist.most_common(10)
```

## findall
긴 word에서 aeiou를 찾아라

fredist하면 어떤 모음의 철자가 가장 많이 나오는 지 알 수 있음.

str의 처음과 끝에 대한 정의가 없기 때문에, 단어 중간 어디에 나타나도 상관이없다. 

(..)의 의미는?

```python
word = 'supercalifragilisticexpialidocious'
result = re.findall(r'[aeiou]', word)
print(result)
result = re.findall('[aeiou](..)[aeiou]', word)
print(result)
result = re.findall('[^aeiou].+[^aeiou]', word)
print(result)
result = re.findall('[^aeiou].+?[^aeiou]', word)
print(result)
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only
cfd.tabulate()
```
